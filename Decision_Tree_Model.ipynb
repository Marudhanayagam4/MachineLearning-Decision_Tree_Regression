```python
import pandas as pd
# data collection
dataset=pd.read_csv("50_Startups.csv")
dataset
```

```python
# get data dummies because of categorical value in state colum
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
# retrieving input  from dataset
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 
       'State_Florida', 'State_New York']]

# retrieving output from dataset
dependent=dataset[['Profit']]
```

```python
# splitting training set and test set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# to train model- use train set -model creation - learning phase

from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion="absolute_error" ,splitter= "best")
regressor.fit(X_train,Y_train) # trained model
DecisionTreeRegressor()
```

```python
# decison tree
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
# after trained the model - prediction
# predicting the output using test set data
Y_predict=regressor.predict(X_test)
```

```python
# to evaluate the trained model using r2 score
# r2 score lies between 0 - 1
# closer to 0 - poor model
# closer to 1 - best model
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_predict)
r_score
```

```python
# use of pickle - to convert data structures, into a format that can be easily stored or transimitted over the network
# pickle in python converting data structure into a byte stream to store it in file/database, maintain program state across sessions or transport data over the network
import pickle
filename="final_Decision_Tree_model.sav"
pickle.dump(regressor,open(filename,"wb")) # save trained best model

## ready to deploy
```
